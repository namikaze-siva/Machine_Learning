```python
#Linear Regression from scratch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
class Linear_Regression:
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.t0=0 #initializing the coefficients
        self.t1=0
    def predict(self,x=0):
        if x==0:
            y_prid=[]
            for i in range(0,len(self.y)):
                Y=self.t0+self.t1*self.x[i]
                y_prid.append(Y)
            return(y_prid)
        else:
            return(self.t0+self.t1*x)
            
    def gradientdescent(self,learning_rate=0.00000001):
        n=len(self.x)
        f1=[]
        f0=[]
        for i in range(0,n):
            f1.append((self.predict()[i]-self.y[i])*self.x[i])
            f0.append(self.predict()[i]-self.y[i])
        self.t1=self.t1-learning_rate*np.mean(f1)
        self.t0=self.t0-learning_rate*np.mean(f0)
    
    def error(self,y_prid):
        n=len(self.y)
        err=[]
        for i in range(0,n):
            err.append(y_prid[i]-self.y[i])
        error=1-np.sum(err)/n
        return(error)
    def costfunc(self,y_prid):
        n=len(self.y)
        err=[]
        for i in range(0,n):
            err.append(y_prid[i]-self.y[i])
            
        return(1/n*(np.sum(err)**2))
    def plot(self,y_prid,name='values'):
        f=plt.figure(name)
        plt.scatter(self.x,self.y,color='blue')
        plt.plot(self.x,y_prid,color='green')
        f.show()
    def get_coefficient(self):
        return((self.t0,self.t1))
```

```python
data=pd.read_csv('Housing.csv')
data.head()
```

```python
x=data['area'].tolist()
x[1:5]
```

```python
y=data['price'].tolist()
y[1:5]
```

```python
def main(x,y):
    model=Linear_Regression(x,y)
    iteration=5000
    for i in range(1,iteration):
        model.gradientdescent()
        if i%50==0:
            error=model.error(model.predict())
            print('for iteration ', i, 'the error is',error)
    y_pred=model.predict()
    print('plot')
    model.plot(y_pred,name='House Rent')
    print('cost function is',model.costfunc(y_pred))
    print('final Accuracy is',model.error(y_pred))
    get_val=model.get_coefficient()
    with open('weight.txt','w') as f:
        f.write('The value of theta0=')
        f.write(str(get_val[0]))
        f.write('\nThe value of theta1=')
        f.write(str(get_val[1]))
        f.close()
        
    
    
    
    
```

```python
if __name__=='__main__':
    main(x,y)
```
